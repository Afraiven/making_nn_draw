```python
import json
import random

file_path = r"C:\Users\Filip\Desktop\learning_to_draw\full_simplified_apple.ndjson"

data_list = []

with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        data = json.loads(line)
        for stroke in data['drawing']:
            if len(stroke[0]) > 5:
                x, y = stroke
                x = x[:6]  
                y = y[:6]
                data_list.append(list(zip(x, y)))  
                break

output_file_path = r"C:\Users\Filip\Desktop\learning_to_draw\output_data.json"
with open(output_file_path, 'w', encoding='utf-8') as output_file:
    json.dump(data_list, output_file)

print(f"Data saved to {output_file_path}")
```

```python
# open saved data and plot
with open(output_file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)
    i = 0
    for drawing in data:
        plt.figure()
        x, y = zip(*drawing)
        plt.plot(x, y, marker='o')
        plt.xlim(0, 255)
        plt.ylim(0, 255)
        plt.gca().invert_yaxis()
        plt.show()
        i += 1
        if i == 3:
            break

```
